```python
#SDO Dataset
sdo_data = '/content/drive/MyDrive/AnkanProjects/AnkanSDOData/aia.lev1.171A_2024-06-19T08_00_09.35Z.image_lev1.fits'
# Verify the existence of the SDO data file
import os
if os.path.isfile(sdo_data):  # Use isfile for individual file validation
    print("SDO data file found:", os.path.basename(sdo_data))
else:
    print("SDO data file not found!")
```

```python
#Necessary imports
import plotly.graph_objects as go
import numpy as np
from astropy.io import fits
```

```python
#Loading FITS Data
fits_data = fits.open(sdo_data)
fits_data.info()
#Accesing Data
image_data = fits_data[0].data
header = fits_data[0].header

print("Data Shape:", image_data.shape)
print("Header Info:", header)

```

```python
import plotly.graph_objects as go

# Create a Plotly figure and add the FITS image
fig = go.Figure()
fig.add_trace(go.Image(z=image_data))

# Customize the layout
fig.update_layout(
    title="FITS Image",
    title_x=0.5,  # Center the title
    xaxis=dict(scaleanchor="y"),  # Maintain aspect ratio
)

# Display the plot
fig.show()
```
